```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.svm import SVC
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,f1_score

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pylab as plt
```

```python
df=pd.read_csv(r'D:\Data_Science\7-Machine_Learning\projects\END-TO-END_projectsML\LOAN_END_TO_END_ML\notebook\data\Loan_after_cleaning&prosessing.csv')
```

```python
df.head()
```

```python
x=df.drop(['Loan_ID','Loan_Status'],axis=1)
y=df['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)

print(x_train.shape)
print(x_test.shape)
```

```python
def evaluate_model(y_true, y_pred):
    f1 = f1_score(y_true, y_pred)
    accuracy = accuracy_score(y_true, y_pred)
    confuse = confusion_matrix(y_true, y_pred)
    report = classification_report(y_true, y_pred)

    return f1, accuracy, confuse, report
```

```python
models={
    "LogisticRegression":LogisticRegression(),
    "tree":DecisionTreeClassifier(),
    "knn":KNeighborsClassifier(),
    "svc":SVC(),
    "XGBClassifier":XGBClassifier(),
    "CatBoostClassifier":CatBoostClassifier(),
    "RandomForest":XGBClassifier(),
    "AdaBoost":AdaBoostClassifier()
}
```

```python
model_list=[]
evaluate_list=[]

for name, model in models.items():
    model.fit(x_train,y_train)

    predict_train=model.predict(x_train)
    predict_test=model.predict(x_test)


    train_f1, train_accuracy, train_confuse, train_report = evaluate_model(y_train, predict_train)
    test_f1, test_accuracy, test_confuse, test_report = evaluate_model(y_test, predict_test)


    print(name)
    model_list.append(name)

    print("model performance from train test")
    print("train_f1-score",train_f1)
    print("train_accuarcy",train_accuracy)
    print("train_confuse",train_confuse)
    print("train_report",train_report)

    print("-----------------------------------------------------------------------")

    print("model performance from test set")
    print("test_f1-score",test_f1)
    print("test_accuarcy",test_accuracy)
    print("test_confuse",test_confuse)
    print("test_report",test_report)
    
    print('='*35)
    print('\n')


    evaluate_list.append(test_f1)

    


result_df=pd.DataFrame({
    'model':model_list,
    'accuracy_score':evaluate_list
})
result_df.head(10)
```

```python
best_model_name=result_df.iloc[0]['model']
best_model_acc=result_df.iloc[0]['accuracy_score']

print("The Best Model Name:",best_model_name)
print("Accuracy score this model is equal = ",best_model_acc)
```
